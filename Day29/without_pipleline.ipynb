```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
# from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
# from sklearn.pipeline import Pipeline
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.drop(columns=["PassengerId","Name","Ticket","Cabin"],inplace=True)
df
```

```python
df.isnull().sum()
# df.head()
```

```python
df.fillna({"Age": df["Age"].mean()}, inplace=True)
```

```python
df.isnull().sum()
df

```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df.drop(columns="Survived"),df['Survived'],test_size=0.2,random_state=42)
X_train.isnull().sum()
```

```python
si_fare=SimpleImputer()
X_train_fare=si_fare.fit_transform(X_train[["Fare"]])
X_test_fare=si_fare.transform(X_test[["Fare"]])
X_train_fare
```

```python
ohe_sex=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked=OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex=ohe_sex.fit_transform(X_train[["Sex"]])
X_train_embarked=ohe_embarked.fit_transform(X_train[["Embarked"]])
print(X_train_sex[3:15,0:2])
print(X_train_sex.shape)
X_test_sex=ohe_sex.transform(X_test[['Sex']])
X_test_embarked=ohe_embarked.transform(X_test[['Embarked']])
print(X_train_embarked.shape)
```

```python
X_train_rem=X_train.drop(columns=['Fare','Sex','Embarked'])
print(X_train_rem.shape)
X_test_rem=X_test.drop(columns=['Fare','Sex','Embarked'])
```

```python
X_train_transformed=np.concatenate((X_train_rem,X_train_fare,X_train_embarked,X_train_sex),axis=1)
X_test_transformed=np.concatenate((X_test_rem,X_test_fare,X_test_embarked,X_test_sex),axis=1)
X_train_transformed
```

```python
clf=DecisionTreeClassifier()
clf.fit(X_train_transformed,Y_train)
```

```python
y_pred=clf.predict(X_test_transformed)
y_pred
```

```python
accuracy=accuracy_score(Y_test,y_pred)
accuracy
```
