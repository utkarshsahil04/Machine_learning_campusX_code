```python
import pickle as pkl
import numpy as np
```

```python
ohe_sex=pkl.load(open('model/ohe_sex.pkl','rb'))
ohe_embarked=pkl.load(open('model/ohe_embarked.pkl','rb'))
clf=pkl.load(open('model/clf.pkl','rb'))
```

```python
test_imput=np.array([2,"male",32.0,0,0,10.5,'S'],dtype=object).reshape(1,7)
test_imput.shape
```

```python
test_input_sex=ohe_sex.transform(test_imput[:,1].reshape(1,1))
test_input_sex.shape
```

```python
test_imput_embarked=ohe_embarked.transform(test_imput[:,6].reshape(1,1))
test_imput_embarked.shape
```

```python
test_input_fare=test_imput[:,5].reshape(1,1)
test_input_fare.shape
```

```python
test_imput_transformed=np.concatenate((test_imput[:,[0,2,3,4]],test_input_sex,test_input_fare,test_imput_embarked),axis=1)
test_imput_transformed.shape
```

```python
clf.predict(test_imput_transformed)
```
