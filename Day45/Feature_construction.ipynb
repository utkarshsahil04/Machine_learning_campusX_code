```python
import numpy as np
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression

import pandas as pd
import seaborn as sns

```

```python
df=pd.read_csv("train.csv",usecols=["Age","Pclass","SibSp","Parch","Survived"])
df.head()
```

```python
df.isnull().sum()
df.dropna(inplace=True)
df.shape
```

```python
df.shape
```

```python
x=df.iloc[:,1:5]
x
y=df.iloc[:,0]
y
```

```python
x.shape
```

```python
cross_val_score(LogisticRegression(),x,y,scoring="accuracy",cv=20).mean()
```

```python
x["Family_size"]=x["SibSp"]+x["Parch"]+1
x
```

```python
def family_class(num):
    if num==1:
        return 0
    elif num>=1 and num<=4:
        return 1
    else :
        return 2
    
```

```python
x["family_type"]=x["Family_size"].apply(family_class)
x
```

```python
x.drop(columns=["SibSp","Parch"],inplace=True)
x.head()
```

```python
x.head()
```

```python
cross_val_score(LogisticRegression(),x,y,scoring="accuracy",cv=20).mean()
```
