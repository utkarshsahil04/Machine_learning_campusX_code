```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,precision_score
import matplotlib.pyplot as plt 
import seaborn as sns
df=pd.read_csv("student.csv")
le=LabelEncoder()
df["Internet"]=le.fit_transform(df["Internet"])
df["Passed"]=le.fit_transform(df["Passed"])
print(df.head())

feature=["StudyHours","Attendance","PastScore","SleepHours"]
std=StandardScaler()
df_scaled=df.copy()
df_scaled[feature]=std.fit_transform(df_scaled[feature])
# df_scaled.head()
#model training part 
model=LogisticRegression()
X=df_scaled[feature]
Y=df_scaled["Passed"]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
model.fit(X_train,Y_train)
y_pred=model.predict(X_test)
# y_pred
accuracy=accuracy_score(Y_test,y_pred)
# print(accuracy)
precision=precision_score(Y_test,y_pred)
# print(precision)
print(classification_report(Y_test,y_pred))
# confusion_mat=confusion_matrix(Y_test,y_pred)

# plt.figure(figsize=(8,6))
# sns.heatmap(confusion_mat,annot=True,fmt='d',cmap="Blues",
#             xticklabels=["Fail","Pass"],yticklabels=["Fail","pass"])
# plt.xlabel("Predicted")
# plt.ylabel("Actual")
# plt.title("confusion Matrix")
# plt.tight_layout()
# plt.show()

try:
    studyhours=int(input("Enter the studyhours"))
    attended_classes=int(input("Enter the Attendence"))
    pastscore=int(input("Enter the Pastscore"))
    sleephours=int(input("Enter the Sleephours"))
    user_given_df=pd.DataFrame([{
        "StudyHours":studyhours,
        "Attendance":attended_classes,
        "PastScore":pastscore,
        "SleepHours":sleephours
    }])
    print(user_given_df)
    user_given_scaled=std.transform(user_given_df)
    print(user_given_scaled)
    prediction=model.predict(user_given_scaled)[0]
    result="Pass" if prediction==1 else "Fail"
    print("prediction IS :",result)
except Exception as e:
    print(f"{e} found")
```
